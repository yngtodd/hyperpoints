```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
run1 = pd.read_csv('progress1.csv')
run2 = pd.read_csv('progress2.csv')
run3 = pd.read_csv('progress3.csv')
run4 = pd.read_csv('progress4.csv')
run5 = pd.read_csv('progress5.csv')
```

```python
runs = [run1, run2, run3, run4, run5]
rewards = [np.array(x['PMM:episode_rewards']) for x in runs]
df = pd.concat([pd.Series(x) for x in rewards], axis=1)
```

```python
def plot_average_walks(walks):
    """Plot aggregate results of series.
    
    Parameters
    ----------
    walks : pd.DataFrame
      Dataframe containing random walks as columns.
    """
    mean = np.array(walks.mean(axis=1))
    std_dev = np.array(walks.std(axis=1))
    upper = mean + std_dev
    lower = mean - std_dev
    results = [upper, mean, lower]
    sns.tsplot(results)
```

```python
plot_average_walks(df)
```
